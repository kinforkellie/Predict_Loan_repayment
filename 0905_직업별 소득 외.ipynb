```python
#-*- coding: euc-kr -*-

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from pandas import DataFrame, Series

%matplotlib inline

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path = 'challenge_data/'

data = pd.read_csv(path + 'Data_set.csv', encoding = 'euc-kr')
test = pd.read_csv(path + 'Test_set.csv', encoding = 'euc-kr')
```

```python
data['OCCP_NAME_G'] = data['OCCP_NAME_G'].replace('*', 'missing') 

data['OCCP_NAME_G'].replace(' ', '', regex=True, inplace=True)
```

```python
print(data.OCCP_NAME_G.isnull().sum())
print(data.CUST_JOB_INCM.isnull().sum())
print(data.HSHD_INFR_INCM.isnull().sum())
print(data.ACTL_FMLY_NUM.isnull().sum())
```

```python
data['OCCP_NAME_G'].fillna('missing', inplace=True)
```

```python
data['MATE_OCCP_NAME_G'] = data['MATE_OCCP_NAME_G'].replace('*', 'missing') 
```

```python
data['MATE_OCCP_NAME_G'] = data['MATE_OCCP_NAME_G'].replace(np.nan, '-') 
```

```python
data['MATE_OCCP_NAME_G'] = data['MATE_OCCP_NAME_G'].replace(' ', '', regex=True) 
```

```python
# 직업별 평균 소득
occp_incm = pd.DataFrame()
occp_incm['mean'] = data.groupby('OCCP_NAME_G')['CUST_JOB_INCM'].mean().round(2)

# 직업별 가구수 / 배우자 추정 소득 / 신용대출건수 / 신용대출금액 / 신용상환금액 / 최초신용등륵 / 최근신용등급 / 연체율
occp_incm['mean_fmly_num'] = data.groupby('OCCP_NAME_G')['ACTL_FMLY_NUM'].mean().round(2)
occp_incm['mean_mate_incm'] = data.groupby('OCCP_NAME_G')['MATE_JOB_INCM'].mean().round(2)
occp_incm['mean_loan_cnt'] = data.groupby('OCCP_NAME_G')['CRDT_LOAN_CNT'].mean().round(2)
occp_incm['mean_crnl_amt'] = data.groupby('OCCP_NAME_G')['TOT_CRLN_AMT'].mean().round(2)
occp_incm['mean_crdt_amt'] = data.groupby('OCCP_NAME_G')['TOT_REPY_AMT'].mean().round(2)
occp_incm['mean_balance_rate'] = (1 - (occp_incm['mean_crdt_amt'] / occp_incm['mean_crnl_amt'])) * 100
occp_incm['mean_start_crdt'] = data.groupby('OCCP_NAME_G')['STRT_CRDT_GRAD'].mean().round(2)
occp_incm['mean_ltst_crdt'] = data.groupby('OCCP_NAME_G')['LTST_CRDT_GRAD'].mean().round(2)
occp_incm['mean_ovdu_rate'] = data.groupby('OCCP_NAME_G')['PREM_OVDU_RATE'].mean().round(2)
```
